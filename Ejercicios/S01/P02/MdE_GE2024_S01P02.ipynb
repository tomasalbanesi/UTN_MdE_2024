## 2. Inicializaciones de código <a class="anchor" id="2"></a>

```python
import numpy as np
import matplotlib.pyplot as plt

# Constantes
k = 8.9875517923 * 10**9  # Constante de Coulomb en N*m^2/C^2
q = 1.0  # Carga total del anillo en C
b = 1.0  # Radio del anillo en metros
epsilon_0 = 8.854187817e-12  # Permitividad del vacío en F/m
```

## 3. Resolución analítica <a class="anchor" id="3"></a>

## 4. Resolución numérica <a class="anchor" id="4"></a>

```python
# Función para calcular el campo eléctrico en el eje z generado por el anillo en un punto z
def campo_electrico_anillo(z):   
    factor = q / (4 * np.pi * epsilon_0)
    return factor * (z / (np.sqrt(b**2 + z**2))**3)

# Valores de z para evaluar el campo eléctrico
z_values = np.linspace(0, 5, 1000)

# Calcular el campo eléctrico en función de z
campo_electrico = campo_electrico_anillo(z_values)

# Graficar el campo eléctrico en función de z
plt.plot(z_values, campo_electrico)
plt.xlabel('Posición en el eje z (m)')
plt.ylabel('Campo eléctrico (N/C)')
plt.title('Campo eléctrico de un anillo con distribución lineal de carga, sobre punto del eje Z y radio b constante')
plt.grid(linewidth=0.5, alpha=0.7)
plt.show()
```
